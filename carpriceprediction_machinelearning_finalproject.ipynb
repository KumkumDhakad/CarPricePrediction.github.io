```python
X = car_data.drop(["Car_Name" , "Selling_Price"] , axis=1)
```

```python
X
```

```python
y = car_data["Selling_Price"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.20 , random_state = 42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()
gbr.fit(X_train , y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfg = RandomForestRegressor()
rfg.fit(X_train , y_train)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
etr = ExtraTreesRegressor()
etr.fit(X_train , y_train)
```

```python
from xgboost import XGBRegressor
xgbr = XGBRegressor()
xgbr.fit(X_train , y_train)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train , y_train)
```

```python
y_predict1 = gbr.predict(X_test)

y_predict2 = rfg.predict(X_test)

y_predict3 = etr.predict(X_test)

y_predict4 = xgbr.predict(X_test)

y_predict5 = lr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test , y_predict1)

score2 = metrics.r2_score(y_test , y_predict2)

score3 = metrics.r2_score(y_test , y_predict3)

score4 = metrics.r2_score(y_test , y_predict4)

score5 = metrics.r2_score(y_test , y_predict5)
```

```python
print(score1 , score2 , score3 , score4 , score5)
```

```python
final_data = pd.DataFrame({'Models':['GBR','RFG','ETR','XGBR','LR'],
             "R2_SCORE":[score1,score2,score3,score4,score5]})
```

```python
final_data
```

```python
sns.barplot(x=final_data['Models'], y=final_data['R2_SCORE'])
```
